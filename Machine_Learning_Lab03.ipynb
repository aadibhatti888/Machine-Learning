**Generating Random Vector**

```python
import random
import numpy as np  

u=[]
for i in range(1,3):
    u.append(random.randrange(10))
```

```python
u
```

```python
type(u)
```

```python
# type(rand_vec)
# rand_vec[1]=-rand_vec[1]
```

```python
u
```

```python
v =np.flip(u)
```

```python
v[1]=-v[1]
```

```python
np.dot(v, np.transpose(u))
```

**Orthogonal Vector**

```python
v
```

**Projection of Random Vector on Orthogonal Vector**

```python
v_norm = np.sqrt(sum(v**2))    

proj_of_u_on_v = (np.dot(u, v) / v_norm**2)*v
```

```python
print(proj_of_u_on_v)
```

**Projetion of Orthogonal Vector on Random Vector**

```python
u_norm = np.sqrt(sum(u**2))    

proj_of_v_on_u = (np.dot(v, u) / u_norm**2)*u
```

```python
print(proj_of_u_on_v)
```

**Two randomly selected vectors projected one on the other** 

```python
u=[]
v=[]
for i in range(1,3):
    u.append(random.randrange(10))
    v.append(random.randrange(10))
```

```python
u=np.array(u)
v=np.array(v)
```

```python
v_norm = np.sqrt(sum(v**2))    
u=np.transpose(u)
proj_of_u_on_v = (np.dot(u, v) / v_norm**2)*v
```

```python
proj_of_u_on_v
```
