```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\CITY COMPUTERS\Downloads\ECG200_TRAIN.csv', header=None , delimiter="  ")
```

```python
dfs_List = []
df = df.sort_values(by=[0])
```

```python
dataframe1 = df.head(25)
dataframe2 = df.tail(75)
```

```python
# Indices
df1_indices = 0
df2_indices = 0

# ietrating till 7 folds
for i in range(0, 7):
    rand_df1 = dataframe1.iloc[df1_indices : df1_indices + 6]
    rand_df2 = dataframe2.iloc[df2_indices : df2_indices + 14]
    
    df2_indices =df2_indices+7
    df1_indices =df1_indices+3
    # concatenation
    rand_dfs_after_concat = pd.concat([rand_df1, rand_df2])
    # 
    dfs_List.append(rand_dfs_after_concat)
```

```python
# Seven folds :/ value of k is 7 in k-fold
fold1 = dfs_List[0]
fold2 = dfs_List[1]
fold3 = dfs_List[2]
fold4 = dfs_List[3]
fold5 = dfs_List[4]
fold6 = dfs_List[5]
fold7 = dfs_List[6]

# printing 7th fold
fold7
```

```python
# Concatenating folds for train and test-set

# fold-1
X_train1 = pd.concat([fold1, fold2, fold3, fold4, fold5, fold6])
y_test1 = fold7

# fold-2
X_train2 = pd.concat([fold1, fold2, fold3, fold4, fold5, fold7])
y_test2 = fold6

# fold-3
X_train3 = pd.concat([fold1, fold2, fold3, fold4, fold6, fold7])
y_test3= fold5

# fold-4
X_train4 = pd.concat([fold1, fold2, fold3, fold5, fold6, fold7])
y_test4 = fold4

# fold-5
X_train5 = pd.concat([fold1, fold2, fold4, fold5, fold6, fold7])
y_test5= fold3

# fold-6
X_train6 = pd.concat([fold1, fold3, fold4, fold5, fold6, fold7])
y_test6= fold2

# fold-7
X_train7 = pd.concat([fold2, fold3, fold4, fold5, fold6, fold7])
y_test7= fold1

# concatenating train and test-sets
train_set = [X_train1, X_train2, X_train3, X_train4, X_train5, X_train6, X_train7]
test_set  = [y_test1, y_test2, y_test3, y_test4, y_test5, y_test6, y_test7 ]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# K-nearest Neighbour
#model = KNeighborsClassifier(n_neighbors=3)


# Decision tree
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
# Loop ietrating till 7-folds
accuracies=[]
for i in range(0, 7):
    # training set
    ytrain = train_set[i]
    ytrain = ytrain.iloc[: , 0]
    ytrain = ytrain.to_numpy()
    ytrain = np.where(ytrain == 1.0, 1 , ytrain)
    ytrain = np.where(ytrain == -1.0 ,-1 , ytrain)
    # xtrain
    xtrain = train_set[i]
    xtrain = xtrain.loc[: , 1:98]
    xtrain = xtrain.to_numpy()

    # fitting model
    clf.fit(xtrain , ytrain)
    # testing set
    
    #ytest
    ytest = test_set[i]
    ytest = ytest.iloc[: , 0]
    ytest = ytest.to_numpy()
    ytest = np.where(ytest == -1.0 , -1 , ytest)
    ytest = np.where(ytest == 1.0 , 1 , ytest)
    # xtest
    xtest = test_set[i]
    xtest = xtest.loc[: , 1:98]
    xtest = xtest.to_numpy()
    
    # Model Prediction
    Ypred = clf.predict(xtest)
    
#     print(y_pred)
#     print(Y_test)
    print("\n******************************************\n Fold" , i,"\n")
    # finding accuracy scores and appending them to a list of accuracies
    from sklearn.metrics import confusion_matrix
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import f1_score
    from sklearn.metrics import classification_report 
    print ("Accuracy : " , accuracy_score(ytest,Ypred)*100)  
    print("Report : \n", classification_report(ytest, Ypred))
    print("F1 Score : ",f1_score(ytest, Ypred, average='macro')*100)
    
    
    accur = accuracy_score(ytest,Ypred)
    accuracies.append(accur*100)
```

```python
array_of_accuracies = np.array(accuracies).astype(int)
array_of_accuracies=array_of_accuracies.tolist()
```

```python
for i in range(len(array_of_accuracies)):
    print("Fold",i+1," : ",array_of_accuracies[i], "%")
```
